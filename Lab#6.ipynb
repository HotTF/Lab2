```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

# Exercise 1: Read Total profit of all months and show it using a line plot
# Total profit data provided for each month. Generated line plot must include the following properties: –
# X label name = Month Number
# Y label name = Total profit
print('1:')
sales=pd.read_csv('company_sales_data.csv')
months = sales['month_number']
total_profit = sales['total_profit']
plt.figure(figsize=(10, 5))
plt.plot(months, total_profit, linestyle='-', color='b')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')
plt.show()

# Exercise 2: Get total profit of all months and show line plot with the following Style properties
# Generated line plot must include following Style properties: –
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3
print('2:')
months = sales['month_number']
total_profit = sales['total_profit']
plt.figure(figsize=(10, 5))
plt.plot(months, total_profit, marker='o', linestyle=':', color='r', 
         markerfacecolor='red', linewidth=3, label='Profit data of last year')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Company Sales Data')
plt.legend(loc='lower right')
plt.show()

 
# Exercise 3: Read all product sales data and show it  using a multiline plot
# Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
print('3:')
months = sales['month_number']
facecream = sales['facecream']
facewash = sales['facewash']
toothpaste = sales['toothpaste']
bathingsoap = sales['bathingsoap']
shampoo = sales['shampoo']
moisturizer = sales['moisturizer']
plt.figure(figsize=(10, 6))
plt.plot(months, facecream, marker='o', linestyle='-', label='Face Cream Sales')
plt.plot(months, facewash, marker='o', linestyle='-', label='Face Wash Sales')
plt.plot(months, toothpaste, marker='o', linestyle='-', label='Toothpaste Sales')
plt.plot(months, bathingsoap, marker='o', linestyle='-', label='Bathing Soap Sales')
plt.plot(months, shampoo, marker='o', linestyle='-', label='Shampoo Sales')
plt.plot(months, moisturizer, marker='o', linestyle='-', label='Moisturizer Sales')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Sales Data for Each Product')
plt.legend(loc='upper left')
plt.show()

# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
# Also, add a grid in the plot. gridline style should “–“.
print('4:')
months = sales['month_number']
toothpaste_sales = sales['toothpaste']
plt.figure(figsize=(10, 6))
plt.scatter(months, toothpaste_sales, color='b', label='Toothpaste Sales')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Units Sold')
plt.title('Toothpaste Sales Data')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--')
plt.show()

# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
# The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.
print('5:')
months = sales['month_number']
facecream_sales = sales['facecream']
facewash_sales = sales['facewash']
bar_width = 0.35
r1 = range(len(months))
r2 = [x + bar_width for x in r1]
plt.figure(figsize=(10, 6))
plt.bar(r1, facecream_sales, color='b', width=bar_width, edgecolor='grey', label='Face Cream Sales')
plt.bar(r2, facewash_sales, color='r', width=bar_width, edgecolor='grey', label='Face Wash Sales')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Sales Data')
plt.xticks([r + bar_width / 2 for r in range(len(months))], months)
plt.legend()
plt.show()

# Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
print('6:')
sales = pd.read_csv('company_sales_data.csv')
bathing_soap_sales = sales['bathingsoap']
plt.figure(figsize=(10, 6))
plt.bar(sales['month_number'], bathing_soap_sales, color='skyblue')
plt.title('Bathing Soap Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.xticks(sales['month_number'])
plt.grid(True)
plt.savefig('bathing_soap_sales_bar_chart.png')
plt.show()

# Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
print('7:')
sales = pd.read_csv('company_sales_data.csv')
total_profit = sales['total_profit']
plt.figure(figsize=(10, 6))
plt.hist(total_profit, bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Total Profit')
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
# Note: In Pie chart display Number of units sold per year for each product in percentage.
print('8:')
sales = pd.read_csv('company_sales_data.csv')
last_year_sales = sales.tail(12)
total_units_last_year = last_year_sales.sum()[1:7]
product_names = total_units_last_year.index
plt.figure(figsize=(10, 6))
plt.pie(total_units_last_year, labels=product_names, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Units Sold per Product in Last Year')
plt.axis('equal') 
plt.show()

# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
print('9:')
sales = pd.read_csv('company_sales_data.csv')
bathing_soap = sales['bathingsoap']
facewash = sales['facewash']
months = sales['month_number']
plt.subplot(1, 2, 1)
plt.plot(months, bathing_soap, marker='o', color='blue')
plt.title('Bathing Soap Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.subplot(1, 2, 2)
plt.plot(months, facewash, marker='o', color='green')
plt.title('Facewash Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()

# Exercise Question 10: Read all product sales data and show it using the stack plot
print('10:')
sales = pd.read_csv('company_sales_data.csv')
facecream = sales['facecream']
facewash = sales['facewash']
toothpaste = sales['toothpaste']
bathingsoap = sales['bathingsoap']
shampoo = sales['shampoo']
moisturizer = sales['moisturizer']
months = sales['month_number']
plt.figure(figsize=(10, 6))
plt.stackplot(months, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer, labels=['Facecream', 'Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.show()

print('finish')
```
